# 샘플 데이터 생성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=400, n_features=2, n_classes=2,n_redundant=0,random_state=42)
```

```python
print(x[1],y[1])
```

```python
plt.figure(figsize=(3.2,3))
plt.scatter(x[:,0],x[:,1],c=y)
plt.grid(True)
plt.show()
```

# 상관분석

```python
import seaborn as sns
```

```python
xy = np.zeros(shape=(400,3))
xy[:,0] = x[:,0]
xy[:,1] = x[:,1]
xy[:,2] = y
xy[0]
```

```python
data = pd.DataFrame(xy, columns = ['x0','x1','y'])
```

```python
sns.pairplot(data,hue='y',height=2)
plt.show()
```

```python
data.corr()
```

```python
sns.pairplot(data,height=1.5)
plt.show()
```

# 분류값 예측

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### train데이터와 test데이터로 분리

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=42)
```

### 로지스틱 회귀모형 생성

```python
model = LogisticRegression(random_state=42)
```

### 모형 학습

```python
model.fit(train_x,train_y)
```

### 분류값 예측

```python
pred_y = model.predict(test_x)
print(pred_y)
```
