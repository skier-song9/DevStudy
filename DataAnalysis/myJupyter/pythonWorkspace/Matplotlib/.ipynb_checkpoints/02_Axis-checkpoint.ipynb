# 2. 그래프 축

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
```

### 개별개체 폰트 설정 : fontdict 속성

```python
plt.plot(x,y)
plt.title('꺾은선 그래프',fontdict={'family':'HYGungSo-Bold','size':20})
# 개별 폰트 설정
plt.show()
```

### 축 라벨 설정 : plt.xlabel() , plt.ylabel()
### 축 라벨 글자색 : color 속성

```python
plt.plot(x,y)
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='#00aa00')
```

### 축 라벨 위치 조정 : loc 속성
* x축 : left,right,center
* y축 : tom, bottom, center

```python
plt.plot(x,y)
plt.xlabel('X축',color='red',loc='right') # left, right, center 3가지 값 전달 가능
plt.ylabel('Y축',color='#00aa00',loc='top') # top,center,bottom 가능
```

### x,y축의 값 지정 : plt.xticks(리스트) plt.yticks(리스트)

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
