# 5. 파일 저장

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
x = [1,2,3]
y = [2,4,8]
```

## plt.savefig('파일명', dpi = n)

특정 dpi 로 해당 파일명으로 저장.
plot.figure(dpi)를 했을 때는 주피터 노트북에서 볼 때의 dpi를 지정하는 것이고, savefig에서는 파일 저장 시 dpi를 지정하는 것이므로 서로 다르다.

```python
plt.plot(x,y)
plt.savefig('graph.png',dpi = 100)
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_200.png')
```
