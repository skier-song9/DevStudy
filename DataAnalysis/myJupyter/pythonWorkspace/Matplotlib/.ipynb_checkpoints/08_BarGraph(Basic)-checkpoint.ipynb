# 8. 막대그래프 (기본)

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

## plt.bar(x축데이터,y축데이터)

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184] 

plt.bar(labels,values,color='g')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184] 
colors = ['r','g','b'] 

plt.bar(labels,values,color=colors,alpha=0.5) # 각 막대의 색을 지정
```

### plt.ylim(min,max)  : y축 데이터를 제한

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184] 

plt.bar(labels,values)
plt.ylim(175,195) # y축의 데이터를 제한
```

```python
plt.bar(labels,values,width=0.5)
plt.ylim(170,200)
```

```python
plt.bar(labels,values,width=0.5)
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 이름 데이터 각도를 45도로 설정
plt.show()
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
ticks = ['1번학생','2번학생','3번학생']

plt.bar(labels,values)
plt.xticks(labels,ticks)
plt.show()
```
