# 3. 범례 (legend)

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
x = [1,2,3]
y = [2,4,8]
```

## plot의 label 속성 + plt.legend() 함수

```python
plt.plot(x,y,label='데이터에 대한 설명')
plt.legend()
```

```python
plt.plot(x,y,label='데이터에 대한 설명')
plt.legend(loc='upper right')
# upper, lower, right, left, ...
# 구글에서 matplotlib docs 참고 : https://matplotlib.org/3.7.1/api/legend_api.html
```

```python
plt.plot(x,y,label='범례')
plt.legend(loc=(0.7,0.8)) # x축, y축
```
