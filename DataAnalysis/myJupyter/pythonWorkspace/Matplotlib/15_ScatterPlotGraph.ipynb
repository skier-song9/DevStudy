```python
import pandas as pd
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

## plt.scatter(x,y)

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes) 
# s 키워드 인자 : 점의 크기 설정
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 200
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis',alpha=0.6) 
# c : 색을 구분할 기준
# cmap : 지정할 색상 >> matplotlib cmap 검색하면 나옴.
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize = (6,5))

plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis',alpha=0.6) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1,2,3], label='학년',shrink=0.7, orientation='horizontal')
plt.show()
```
