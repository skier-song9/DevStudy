# 14 원 그래프 (심화)

```python
import pandas as pd
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

```python
values=[30,25,20,13,10,2]
labels=['python','java','javascript','c#','c/c++','etc']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05]*6

plt.pie(values,labels=labels, autopct='%.1f%%',startangle=90,colors=colors,explode=explode)
plt.show()
```

```python
wedgeprops = {'width':0.5,'edgecolor':'w','linewidth':2}
plt.pie(values,labels=labels, autopct='%.1f%%',startangle=90,colors=colors,wedgeprops=wedgeprops)
plt.show()

# wedgeprops : 각 파이의 속성을 설정
#   width : 가운데를 뚫어줌 >> 도넛 모양
#   edgecolor : 파이의 테두리 색상
#   edgewidth : 파이 테두리의 두께
```

### 키워드 인자로 함수를 전달하면 각 데이터에 대해 함수를 적용시킨다.

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >= 10 else ''

# pctdistance : text를 원의 중심으로부터 설정 값만큼 멀어지게 한다.
plt.pie(values,labels=labels, autopct=custom_autopct,startangle=90,colors=colors,wedgeprops=wedgeprops,pctdistance=0.7)
plt.show()
```
