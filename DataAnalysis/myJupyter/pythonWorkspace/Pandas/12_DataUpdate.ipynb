# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_csv('score.csv',index_col='지원번호')
df
```

## Column 수정

### series객체.replace({기존값:대체할값})

```python
df['학교'].replace({'북산고':'상북고'}) # 북산고를 상북고로 수정
```

```python
df['학교'].replace({'북산고':'상북고','능남고':'대일고'}) # 북산고를 상북고로 수정
```

```python
df['학교'].replace({'북산고':'상북고'},inplace=True)
```

```python
df
```

### 할당 연산자 = 사용

```python
df['SW특기'] = df['SW특기'].str.lower() # SW특기를 소문자로 수정
df
```

```python
df['학교'] += '등학교'
df
```

## Column 추가

### df객체[새로운 컬럼명] = 데이터

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 컬럼을 추가하고 데이터를 Fail로 초기화
df
```

```python
df.loc[df['총합']>400,'결과'] = 'Pass'
# 총합이 400점이 넘는 행에 대해 '결과'컬럼의 값을 Pass로 수정
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 컬럼을 삭제
```

```python
df.drop(columns=['국어','영어','수학']) # 국영수 컬럼을 삭제
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터를 삭제
```

```python
filt = df['수학']<80 #수학점수가 80점 미만인 데이터
ind = df[filt].index # 필터에 적용되는 학생들의 'index'를 구함
df.drop(index=ind)
```

## Row 추가

```python
df.loc['9번'] = ['이정환','해남고등학교',176,90,90,90,90,90,'Kotlin',450,'Pass']
# 새로운 row 추가
df
```

## 특정 cell 수정

```python
df.loc[df['이름']=='서태웅','SW특기'] = 'Python'
# 4번 학생의 SW특기 데이터를 python으로 수정
df
```

```python
df.loc[df['이름']=='강백호',['학교','SW특기']] = ['하나고등학교','C++']
# 강백호 학생의 학교, SW특기 데이터 수정
df
```

## Column 순서 변경

### 파이썬 list를 활용하여 컬럼 순서 변경

```python
cols = list(df.columns) # 파이썬 list타입으로 형변환
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]]
# 맨 뒤의 컬럼을 앞으로 가져오고, 나머지 컬럼들을 뒤에 나열
df
```

## Column label 변경

### df객체.columns = [변경할 이름]

```python
cols = list(df.columns)
df.columns = cols[:1] + ['성함'] + cols[2:]
df
```
