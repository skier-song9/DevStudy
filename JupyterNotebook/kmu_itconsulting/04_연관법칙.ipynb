# example.csv파일로부터 데이터 읽기

```python
data = pd.read_csv("./data/example.csv")
data
```

```python
data=data.fillna(0)
data2=data.drop("trainID",axis=1)
data2
```

# 연관 법칙 분석 함수 import
# 연관 항목 쌍 수집

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
minSupportData = apriori(data2,min_support=0.2,use_colnames=True)
minSupportData
```

```python
minSupportData04 = apriori(data2,min_support=0.4,use_colnames=True)
minSupportData04
```

# 연관 법칙 추출

```python
example_Rules_pos = association_rules(minSupportData04,metric="lift",min_threshold=1)
example_Rules_pos
```

```python
example_Rules_pos.sort_values(by=["lift"],ascending=False)
```
