# 상관분석

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
w = pd.read_csv('./data/chi_weight.txt')
w.head(3)
```

## df객체.corr() : 각 열 간의 상관 계수를 반환하는 함수

```python
w.corr()
# 해당 df객체에서 숫자에 해당하는 열들 간의 상관 계수를 반환한다.
```

```python
import seaborn as sns
# seaborn : matplotlib을 기반으로 만들어진 라이브러리로 high-level interface를 제공
```

```python
sns.heatmap(w.corr(), annot=True, cmap='Accent')
plt.show()
```

## sns.pairplot(), plt.scatter() : scatter graph 그리기 

```python
sns.pairplot(w[['weight','egg_weight','movement','food']])
plt.show()
```

#### >> 산포도 그래프로 보았을 때, 병아리의 몸무게는 달걀의 무게, 병아리 먹이의 양과 양의 상관관계를 가지는 사실을 알 수 있다.
