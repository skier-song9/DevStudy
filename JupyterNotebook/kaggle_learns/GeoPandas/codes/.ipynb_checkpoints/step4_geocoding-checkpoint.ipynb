```python
def my_geocoder(row):
    try:
        name = row.Name
        location = geolocator.geocode(name)
        coord = [location.point.latitude,location.point.longitude]
        return pd.Series({'Latitude':coord[0],'Longitude':coord[1]})
    except:
        return None
# universities[['Latitude','Longitude']] = 
universities.apply(my_geocoder,axis=1)
# print("{}% of addresses were geocoded".format((1-sum(pd.isnull(universities["Latitude"]))/len(universities))*100))

# Drop universities that were not geocoded
# universities = universities.dropna(subset=['Latitude','Longitude'])

```

```python
universities = gpd.GeoDataFrame(
    universities, geometry=gpd.points_from_xy(universities.Longitude, universities.Latitude))
universities.set_crs(epsg=4326)
universities.head()
```

```python
london=geolocator.geocode("London")

m = folium.Map(location=[london.point.latitude,london.point.longitude],titles='openstreetmap',zoom_start=3)

for idx,row in universities.iterrows():
    Marker(location=[row.Latitude,row.Longitude],popup=row.Name).add_to(m)
m
```
