# GeoCoder 사용해서 번지 주소를 위경도로 변환

```python
data = pd.read_csv("../data/지번_서울특별시.csv",header=0,encoding="cp949")
data.head(3)
```

```python
data = data.drop(columns=['관리번호','일련번호','법정동코드','법정리명','대표여부'])
data.head(3)
```

```python
gdata = data[data["시군구명"]=="광진구"]
print(len(gdata))
gdata = gdata[gdata["산여부"]==0] # 산 제거
gdata = gdata.reset_index().drop(columns="index",axis=1)
gdata.head(3)
```

```python
len(gdata)
```

```python
gaddr_list = list()
for row in gdata.values:
    gaddr_list.append("{} {} {} {}{}".format(
        row[0],row[1],row[2],row[4],'-'+str(row[5]) if row[5]!=0 else ''))   
#     print("{} {} {} {}{}".format(
#         row[0],row[1],row[2],row[4],'-'+str(row[5]) if row[5]!=0 else ''))
```

```python
regex = r'서울특별시 광진구 \S*동 [\d|-]*'
pattern = re.compile(regex)
```

```python
# detail_addr_index = data_list[0].columns.get_loc('대지위치')
# print(detail_addr_index)
```

```python
locInfo = list()
err_loc = list()
for index,e in enumerate(gaddr_list):
        encoded_addr = quote_plus(e)
        url = "http://api.vworld.kr/req/address?service=address&request=getcoord&version=2.0&crs=epsg:4326&address={}&refine=true&simple=false&format=json&type=parcel&key=07CCF2CD-23F5-3D13-BF3D-1F7676664B6A".format(encoded_addr)
        try:
            request = Request(url)
            response = urlopen(request)
            res_code = response.getcode()
            if res_code == 200:
                response_body = response.read().decode('utf-8')
                jsonData = json.loads(response_body)
#                     pprint(jsonData)
                lat = float(jsonData['response']['result']['point']['x'])
                lng = float(jsonData['response']['result']['point']['y'])
                locInfo.append({"bunji":e,"lat":lat,"lng":lng})
            else:
                print('API error')
        except Exception:
           err_loc.append(e) 
```
