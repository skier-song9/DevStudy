## 데이터 제외하기(삭제하기) : dropna()

```python
df = pd.read_csv('score.csv',index_col='지원번호')
df.dropna() # NaN을 포함하는 행을 삭제
```

### dropna()의 키워드 인자
* axis : index(row를 삭제,default) , columns(컬럼을 삭제)
* how : any(하나라도 있으면 삭제,default) , all(모두 NaN일 때 삭제) 

```python
df.dropna(axis='index',how='any') # NaN이 '하나라도 있는', 'row'를 삭제
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 컬럼을 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns',how='all') #데이터 전체가 NaN인 경우만 컬럼 삭제
```
