# 3. Index

정수 위치 인덱스 = 데이터에 접근할 수 있는 주소 값.
인덱스 이름 = 임의로 지정한 인덱스 값.

```python
import pandas as pd
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

```python
df.index
```

## Index 이름의 컬럼명설정 : df객체.index.name

```python
df.index.name = "지원번호"
df
```

```python
# df['지원번호'] # [x] 인덱스의 컬럼명으로는 접근 불가
```

## Index 초기화 : 데이터프레임객체.reset_index()

### reset_index() : 인덱스를 초기화하고, 설정했던 인덱스 이름은 하나의 컬럼으로 만든다.

```python
df.reset_index()
```

### reset_index(drop=True) : 설정했던 인덱스를 버리고 초기화시킨다.

```python
df.reset_index(drop=True)
```

### reset_index(inplace=True) : 변경된 내용을 실제 데이터에 반영한다.

inplace 위치 인자를 전달하지 않으면 원본 데이터가 변형되지 않는다.

```python
df.reset_index(drop=True,inplace=True)
df
```

## Index 설정 : df객체.set_index()

지정한 column으로 인덱스 설정

```python
df.set_index('이름',inplace=True)
```

## Index를 기준으로 정렬 : df객체.sort_index()

```python
df.sort_index() # 인덱스를 기준으로 오름차순 정렬
```

```python
df.sort_index(ascending=False) #내립차순
```
