# 9. 데이터 선택 (조건에 따라)

```python
import pandas as pd
df = pd.read_csv('score.csv',index_col='지원번호')
df
```

## df객체[조건식]

```python
df['키'] >= 185 # 학생들의 키가 185이상인지 여부를 반환함.
```

```python
filter_ = (df['키'] >= 185)
df[filter_]
```

```python
df[~filter_] # 조건식에 Not을 붙임. ~ = not 연산자
```

```python
df[df['키'] >= 185] # 조건식을 바로 전달해도 된다.
```

## loc, iloc에 조건식 활용

row select에 조건식을 전달하면, 1차적으로 행을 필터링하고 그다음에 필요한 열 데이터만 추출할 수 있다.

```python
df.loc[filter_,'수학'] # 키가 185이상인 학생들의 수학 데이터
```

## 다양한 조건

### & 연산자 = 그리고, and

```python
filt_and = (df['키']>=185) & (df['학교']=='북산고')
df.loc[filt_and] # 키가 185이상'이고' 학교가 북산고인 학생 데이터
```

### | 연산자 = 또는 = or

```python
filt_or = (df['키']<170) | (df['키']>200)
df.loc[filt_or]
```

### str 함수 활용

```python
filt_str = df['이름'].str.startswith('송') # 송 씨 성을 가진 사람
df[filt_str]
```

```python
filt_str2 = df['이름'].str.contains('태')
df[filt_str2]
```

```python
df[~filt_str2] # 이름에 '태'가 들어가지 않는 사람을 추출
```

### isin(배열) 함수 : 배열에 포함되어 있는지 판단.

```python
langs = ["Python","Java"]
filt_list = df['SW특기'].isin(langs)
df[filt_list]
```

```python
langs_ = ['python','java']
filt_list_ = df['SW특기'].str.lower().isin(langs_)
df[filt_list_]
```

```python
filt_str_ = df['SW특기'].str.lower().str.startswith('java',na=False)
# 🔔NA, NaN 데이터에 대해서는 모두 False로 설정하는 인자.
df[filt_str_]
```
