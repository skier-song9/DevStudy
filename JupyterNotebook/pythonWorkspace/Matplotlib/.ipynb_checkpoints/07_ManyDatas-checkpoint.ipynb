## 코로나 백신 종류별 접종 인구

```python
days = [1,2,3] # 1일, 2일, 3일
az = [2,4,8] # 1일부터 3일까지 아스트라제네카 접종인구 (단위: 만명)
pfize=[5,1,3]
moderna=[1,2,5]
```

```python
plt.plot(days,az)
plt.plot(days,pfize)
plt.plot(days,moderna)
```

```python
plt.plot(days,az, label='az',marker='x')
plt.plot(days,pfize,label='pfizer',marker='o',ls='--')
plt.plot(days,moderna,label='moderna',marker='s',ls=':')
plt.legend(ncol=3) #ncol: legend의 column을 지정
```
