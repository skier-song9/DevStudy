# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

## plt.barh(x,y) : 옆으로 누운 막대 그래프를 그린다.

```python
labels = ['강백호','서태웅','정대만']
values= [190, 187 , 184]

plt.barh(labels,values)
plt.xlim(175,195)
plt.show()
```

## 막대에 무늬 넣기 : bar[index].set_hatch(pattern)

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') 
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels,values)
plt.ylim(175,192.5)

for index, rect in enumerate(bar):
    plt.text(index,rect.get_height()+0.5,values[index],ha='center',color='b')
```
