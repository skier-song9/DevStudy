# 13. 원 그래프 (기본)

```python
import pandas as pd
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
```

## plt.pie(데이터) 

```python
values=[30,25,20,13,10,2]
labels=['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels=labels, autopct='%.1f%%',startangle=90)
# labels = 데이터에 대한 설명
# autopct = 원 그래프의 수치 표시
# startangle = 원그래프를 원하는 각도만큼 회전.
plt.show()
```

## Styles

```python
values=[30,25,20,13,10,2]
labels=['python','java','javascript','c#','c/c++','etc']
explode = [0.2,0.1,0,0,0,0] 

plt.pie(values,labels=labels, autopct='%.1f%%',explode = explode)
# explode = 파이 띄우기 효과 
plt.show()
```

```python
values=[30,25,20,13,10,2]
labels=['python','java','javascript','c#','c/c++','etc']
explode = [0.05] * 6

plt.pie(values,labels=labels, autopct='%.1f%%',explode = explode)
plt.show()
```

```python
plt.pie(values, labels=labels,explode=explode)
plt.legend(loc=(1.2,0.3)) # legend의 x위치를 1.2만큼, y위치를 0.3만큼 이동
plt.title('원 그래프')
plt.show()
```

```python
plt.pie(values, labels=labels,explode=explode)
plt.legend(loc=(1.2,0.3),title='언어') 
plt.show()
```
