# 4. 스타일

```python
import matplotlib.pyplot as plt
# ⚠️ 기본 설정
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 사용 시, 마이너스 부호가 깨지는 현상 해결
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth = 10)
```

## Marker

 https://matplotlib.org/3.7.1/api/markers_api.html#module-matplotlib.markers

```python
plt.plot(x,y,marker='o')
# marker = 데이터를 마커로 표시
```

```python
plt.plot(x,y,linestyle='None',marker='o')
```

```python
plt.plot(x,y,marker='v',markersize=10)
```

```python
plt.plot(x,y,marker='X',markersize=10)
```

```python
plt.plot(x,y,marker='o',markersize=10,markeredgecolor='yellow')
```

```python
plt.plot(x,y,marker='o',markersize=10,markeredgecolor='red',markerfacecolor='blue')
```

## LineStyle

https://matplotlib.org/3.7.1/gallery/lines_bars_and_markers/linestyles.html#linestyles

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')
```

## linecolor

```python
plt.plot(x,y,color='#ff0000')
```

```python
plt.plot(x,y,color='black')
```

```python
plt.plot(x,y,color='g') #색에 대한 약자 있음.
# b= blue, g=green, r=red, y=yellow, w=white, ...
```

## Format

축약 표현. 
color, maker, linestyle을 한 번에 지정

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go')
```

## 축약어

```python
plt.plot(x,y,marker='o',mfc='red',ms=8,mec='y',ls='-.')
# mfc = markerfacecolor, ms = markersize, mec=markeredgecolor ...
```

## 투명도

alpha 속성 : 1에 가까울수록 불투명, 0이면 투명.

```python
plt.plot(x,y,'ro--',alpha=0.3) 
```

## 그래프 크기 : plt.figure()

* figsize 키워드 인자 = (x비율,y비율)
* dpi = 해상도

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.plot(x,y)
```

## 배경색 : plt.figure(facecolor="")

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
