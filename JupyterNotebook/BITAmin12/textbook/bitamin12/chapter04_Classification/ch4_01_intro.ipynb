```python
import matplotlib.pyplot as plt
import numpy as np

# x 값 범위 생성
x = np.linspace(0.001, 1, 100)  # 0에 가까운 값들은 로그 계산이 불가능하므로 0.001부터 시작

# 정보량 함수 계산
I = -np.log2(x)

# 그래프 그리기
plt.figure(figsize=(4, 3))
plt.plot(x, I, label='I(x) = -log2(x)')
plt.xlabel('P(x)')
plt.ylabel('I(x)')
plt.title('Information Content Function')
plt.legend()
plt.grid(True)
plt.show()
```
