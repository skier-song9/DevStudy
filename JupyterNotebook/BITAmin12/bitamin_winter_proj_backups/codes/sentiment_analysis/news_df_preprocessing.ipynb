```python
news_df = pd.read_csv('../../data/scraping/stocknewsapi_rawdata.csv',index_col=0)
```

```python
news_df['date'] = news_df['date'].apply(lambda x : 
                   pd.to_datetime(str(x)[:11]))
print(news_df.shape)
news_df.info()
```

```python
news_df_sorted = news_df.sort_values('date')
news_df_sorted = news_df_sorted.set_index('news_id')
```

```python
# drop duplicates
news_df_dd = news_df_sorted.drop_duplicates()
news_df_dd.shape
```

```python
'''
source_name > labelencoding
topics > labelencoding
    if None > 'Undefined'
    if multiple > then first topic
sentiment > onehot encoding
type > onehot encoding
rank_score
'''
nd = news_df_dd[['source_name',
                  'topics','sentiment','type','rank_score','date']]
nd_ = nd.copy()

### source_name
le = LabelEncoder()
nd['source_name'] = le.fit_transform(nd['source_name'])

### topics
tle = LabelEncoder()
import ast
def string_to_list(string):
    try:
        return ast.literal_eval(string)
    except (SyntaxError, ValueError):
        return [] # if error occurs return empty list
def extract_topic(topic_list):
    if len(topic_list)==0:
        return 'Undefined'
    return topic_list[0]
nd['topics'] = nd['topics'].apply(string_to_list)
nd['topics'] = nd['topics'].apply(extract_topic)
nd['topics'] = tle.fit_transform(nd['topics'])

### sentiment & type
onehot_cols = ['sentiment','type']
oe = OneHotEncoder()
encoded = oe.fit_transform(nd[onehot_cols])
feature_names = oe.get_feature_names_out(onehot_cols)
encoded_dense = encoded.toarray()
#combine
nd.drop(columns=['sentiment','type'],inplace=True)
encoded_df = pd.DataFrame(encoded_dense,
                           columns=feature_names,
                           index=nd.index)
nd = pd.concat([nd,encoded_df],axis=1)
```

```python
### Check for topics distribution
import ast
def string_to_list(string):
    try:
        return ast.literal_eval(string)
    except (SyntaxError, ValueError):
        return [] # if error occurs return empty list
def extract_topic(topic_list):
    if len(topic_list)==0:
        return 'Undefined'
    return topic_list[0]

temp = nd_['topics'].apply(string_to_list)
topic_col = temp.apply(extract_topic)
topic_col.unique()
sns.barplot(topic_col.value_counts(),orient='h')
plt.title('Topics value counts')
# plt.savefig('../../plots/topic_value_counts_for_news.png',dpi=300)
plt.show()
```
