```python
# 압축파일 모두 압축 풀기
filepath_zip = '../data/open.zip'
with ZipFile(filepath_zip, 'r') as zipObj:
    for filename in zipObj.namelist():
        zipRead = zipObj.read(filename)
        globals()[f'{filename}'[:-4]] = pd.read_csv(BytesIO(zipRead))
        print(f'{filename}'[:-4])
```
