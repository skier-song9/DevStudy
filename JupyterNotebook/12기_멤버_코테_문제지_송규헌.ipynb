### 문제 [2-3] - 9점
- df 데이터에서 'body_mass_g'열(column)을 삭제하고 인덱스를 재정렬 한 후 df1 데이터로 저장하여 출력하세요.

<출력 예시>

image.png

```python
df1 = df.drop(axis=1,columns='body_mass_g').reset_index(drop=True)
df1
```

### 문제 [2-4] - 12점
- df1의 'flipper_length_mm'을 아래의 분류 기준에 따라 새로운 변수 'length_cat'를 추가한 후 출력하세요.
- long: 210mm 이상 / middle: 190mm이상 210mm 미만 / short: 190mm 미만

<출력 예시>

image.png

```python
def custom_func(flipper):
    if flipper >= 210 :
        return 'long'
    elif flipper >= 190 :
        return 'middle'
    else:
        return 'short'
df1['length_cat'] = df1['flipper_length_mm'].apply(custom_func)
df1
```

### 문제 [2-5] - 8점
- df1에서 length_cat 열 값의 분포를 시각화하세요.

```python
df1['length_cat'].value_counts().index
```

```python
import matplotlib.pyplot as plt
plt.bar(df1['length_cat'].value_counts().index,df1['length_cat'].value_counts())
plt.show()
```

### 문제 [2-6] - 9점


- df1 데이터에서 성별('sex')이 암컷인 펭귄의 부리 길이('bill_length_mm')의 평균을 구하고 소수점 셋째 자리까지 나타내세요.

### 문제 [2-7] - 10점
