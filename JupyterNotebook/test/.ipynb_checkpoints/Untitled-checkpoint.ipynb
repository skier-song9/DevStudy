```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# x값 범위 설정
x = np.linspace(-10, 10, 1000)

# y값 계산
y = x / (1 - x)

# 데이터프레임 생성
df = pd.DataFrame({'x': x, 'y': y})

# 그래프 그리기
fig, ax = plt.subplots()
ax.plot(df['x'], df['y'], color='blue', linewidth=2.5, linestyle='-', label='y = x / (1 - x)')
ax.legend(loc='upper left', fontsize=12)

ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14,rotation=90)
ax.set_title('y = x / (1 - x)', fontsize=16)
plt.show()
```
